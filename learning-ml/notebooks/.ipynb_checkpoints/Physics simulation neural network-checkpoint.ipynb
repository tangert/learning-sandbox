```python
import pygame
from pygame.locals import *
from pygame.color import *
import pymunk
import pymunk.pygame_util
from pymunk import Vec2d
from pymunk import Body
import math, sys, random
import pandas as pd
import csv
```

```python
# Game scene initialization
```

```python
pygame.init()
screen = pygame.display.set_mode((width, height))
clock = pygame.time.Clock()
running = True
draw_options = pymunk.pygame_util.DrawOptions(screen)
```

```python
width = 1000
height = 600
```

```python
### Physics stuff
space = pymunk.Space()
space.gravity = (0.0, -1500.0)

### walls
static_body = space.static_body

bounding_thickness = 0

boundaries = [
                # bottom
                pymunk.Segment(static_body,
                               (0, 0),
                               (width, 0),
                               bounding_thickness),
                # top
                pymunk.Segment(static_body,
                               (0, height),
                               (width, height),
                               bounding_thickness),
    
                # left
                pymunk.Segment(static_body,
                               (0, 0),
                               (0, height),
                               bounding_thickness),

                # right
                pymunk.Segment(static_body,
                               (width, 0),
                               (width, height),
                               bounding_thickness)
               ]

for line in boundaries:
    line.elasticity = 0.95
    line.friction = 1
    
space.add(boundaries)
```

```python
# SIMULATES JUST RECORDING THE VALUES
# BALL CONFIG
mass = 5
radius = 25
inertia = pymunk.moment_for_circle(mass, 0, radius, (0,0))

simulation_count = 50
all_data = []

for i in range(simulation_count):
    
    body = pymunk.Body(mass, inertia)

    pos_x = random.randint(0,width)
    pos_y = random.randint(0, height)

    force_x = random.randint(-5000, 5000)
    force_y = random.randint(-5000, 5000)

    # apply the initial impulse to get a velocity
    body.position = pos_x, pos_y
    body.body_type = Body.DYNAMIC
    body.apply_impulse_at_world_point((force_x,force_y), body.position)

    shape = pymunk.Circle(body, radius, (0,0))
    shape.elasticity = 0.95
    shape.friction = 10

    # add the shape and body to the world scene
    space.add(body, shape)
    # balls.append(shape)

    ### Update physics
    # 60 FPS
    dt = 1.0/60.0
    
    print("Simulation {}".format(i))

    while abs(body.angular_velocity) > 0.1 or body.angular_velocity == 0:
        
        body.angular_velocity *= 0.975
        space.step(dt)
        
        data = [i,
                body.velocity[0],
                body.velocity[1],
                body.position[0],
                body.position[1]
               ]
        
        all_data.append(data)
        
    space.remove(shape)
    space.remove(body)
        
with open('physics_training.csv','wb') as file:
    
    # set the columns
    file.write("trial_id, vel_x, vel_y, pos_x, pos_y".encode())
    file.write("\n".encode())
    
    # add in the data
    for d in all_data:
        file.write("{},{},{},{},{}".format(d[0],d[1],d[2],d[3],d[4]).encode())
        file.write('\n'.encode())
```
